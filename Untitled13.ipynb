```python
train_dataset = tf.keras.utils.image_dataset_from_directory(
    os.path.join(podaci_dir, "train"),  # Putanja do foldera sa treninzima
    image_size=(224, 224),              # Dimenzije slike
    batch_size=32,                      # Batch size
    shuffle=True                        # Mešanje podataka
)

# Učitaj test skup
test_dataset = tf.keras.utils.image_dataset_from_directory(
    os.path.join(podaci_dir, "test"),   # Putanja do foldera sa test podacima
    image_size=(224, 224),              # Dimenzije slike
    batch_size=32,                      # Batch size
    shuffle=False                       # Nema potrebe za mešanjem podataka
)

# Podela trening skupa na trening i validaciju (90% trening, 10% validacija)
val_split = 0.1
train_batches = tf.data.experimental.cardinality(train_dataset).numpy()  # Evaluiraj TensorFlow tensor i pretvori u integer

# Izračunavanje broja batch-eva za validaciju
val_batches = int(train_batches * val_split)

# Podela podataka na trening i validaciju
val_dataset = train_dataset.take(val_batches)      # Uzimamo 10% za validaciju
train_dataset = train_dataset.skip(val_batches)    # Ostatak ostaje za trening

# Normalizacija (skaliranje piksela na opseg [0, 1])
normalization_layer = tf.keras.layers.Rescaling(1./255)

train_dataset = train_dataset.map(lambda x, y: (normalization_layer(x), y))
val_dataset = val_dataset.map(lambda x, y: (normalization_layer(x), y))
test_dataset = test_dataset.map(lambda x, y: (normalization_layer(x), y))

# Optimizacija performansi (prefetch za paralelno procesiranje)
AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
val_dataset = val_dataset.prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.prefetch(buffer_size=AUTOTUNE)

# Provera veličine skupova
print(f"Trening skup: {tf.data.experimental.cardinality(train_dataset).numpy()} batch-eva")
print(f"Validacioni skup: {tf.data.experimental.cardinality(val_dataset).numpy()} batch-eva")
print(f"Test skup: {tf.data.experimental.cardinality(test_dataset).numpy()} batch-eva")
```

```python
train_dataset = tf.keras.utils.image_dataset_from_directory(
    os.path.join(podaci_dir, "train"),  # Putanja do foldera sa treninzima
    image_size=(224, 224),              # Dimenzije slike
    batch_size=32,                      # Batch size
    shuffle=True                        # Mešanje podataka
)

# Učitaj test skup
test_dataset = tf.keras.utils.image_dataset_from_directory(
    os.path.join(podaci_dir, "test"),   # Putanja do foldera sa test podacima
    image_size=(224, 224),              # Dimenzije slike
    batch_size=32,                      # Batch size
    shuffle=False                       # Nema potrebe za mešanjem podataka
)

# Prvo dobijamo imena klasa
class_names = train_dataset.class_names
print(f"Imena klasa: {class_names}")

# Podela trening skupa na trening i validaciju (90% trening, 10% validacija)
val_split = 0.1
train_batches = tf.data.experimental.cardinality(train_dataset).numpy()  # Evaluiraj TensorFlow tensor i pretvori u integer
val_batches = int(train_batches * val_split)

val_dataset = train_dataset.take(val_batches)      # Uzimamo 10% za validaciju
train_dataset = train_dataset.skip(val_batches)    # Ostatak ostaje za trening

# Normalizacija (skaliranje piksela na opseg [0, 1])
normalization_layer = tf.keras.layers.Rescaling(1./255)

train_dataset = train_dataset.map(lambda x, y: (normalization_layer(x), y))
val_dataset = val_dataset.map(lambda x, y: (normalization_layer(x), y))
test_dataset = test_dataset.map(lambda x, y: (normalization_layer(x), y))

# Optimizacija performansi (prefetch za paralelno procesiranje)
AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
val_dataset = val_dataset.prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.prefetch(buffer_size=AUTOTUNE)

# Provera veličine skupova
print(f"Trening skup: {tf.data.experimental.cardinality(train_dataset).numpy()} batch-eva")
print(f"Validacioni skup: {tf.data.experimental.cardinality(val_dataset).numpy()} batch-eva")
print(f"Test skup: {tf.data.experimental.cardinality(test_dataset).numpy()} batch-eva")

# Kreiranje modela prema tvojoj arhitekturi
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(224, 224, 3)),  # Pretvaranje slike u 1D vektor
    tf.keras.layers.BatchNormalization(),  # Batch normalization
    tf.keras.layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.003)),
    tf.keras.layers.Dense(86, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.002)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='tanh'),
    tf.keras.layers.Dense(16, activation='relu', bias_regularizer=tf.keras.regularizers.l2(0.001)),
    tf.keras.layers.Dense(len(class_names), activation='softmax')  # Softmax za višeklasnu klasifikaciju
])

# Kompilacija modela
model.compile(
    optimizer='adam',                                                       # Optimizator
    loss='sparse_categorical_crossentropy',  # Koristimo sparse categorical crossentropy jer oznake nisu u jedan-hot formatu
    metrics=['accuracy']                                                   # Metrika za evaluaciju
)

# Treniranje modela
history = model.fit(
    train_dataset,  # Trening podaci
    epochs=10,          # Broj epoha (možeš povećati broj epoha)
    validation_data=val_dataset,  # Validacioni skup
    batch_size=32
)
```

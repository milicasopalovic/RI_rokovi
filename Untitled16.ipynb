Ovaj deo koda se koristi za podelu skupa podataka na trening i test skup, kako bi se omogućila obuka i evaluacija modela. Proces je sledeći:

Kreiranje direktorijuma: Prvo se kreiraju direktorijumi za trening (trening) i test skupove (test) unutar ciljnog direktorijuma skup_podataka.
Iteracija kroz klase: Zatim, kroz sve klase u izvornoj lokaciji, proverava se da li je datoteka direktorijum. Za svaku klasu se prikupljaju slike (samo .jpg, .jpeg, .png formati).
Podela na trening i test slike: Za svaku klasu slike se dele u trening i test skupove koristeći train_test_split iz sklearn.model_selection. Test skup čini 30% slika, dok ostatak ide za trening.
Kreiranje poddirektorijuma za klase: U okviru direktorijuma za trening i test, za svaku klasu se kreira odgovarajući poddirektorijum.
Premestanje slika: Slike se zatim premestaju iz izvornog direktorijuma u odgovarajuće direktorijume za trening i test.
Na kraju, ispisuje se broj slika dodeljenih trening i test skupovima za svaku klasu, a nakon završetka podela, ispisuje se poruka da je proces završen i da su podaci organizovani u ciljni direktorijum.

```python
import os
import shutil


izvorni_dir = "./ip2-projekat/materijali"
ciljni_dir = "skup_podataka"

# Kreiranje trening i test direktorijuma
trening_dir = os.path.join(ciljni_dir, "trening")
test_dir = os.path.join(ciljni_dir, "test")
os.makedirs(trening_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

klase = os.listdir(izvorni_dir)

for ime_klase in klase:
    putanja = os.path.join(izvorni_dir, ime_klase)

    if not os.path.isdir(putanja):
        continue

    slike = os.listdir(putanja)
    slike = [os.path.join(putanja, img) for img in slike if img.endswith(('.jpg', '.png', '.jpeg'))]  # Filtrirajte slike

    trening_slike, test_slike = train_test_split(slike, test_size=0.3, random_state=42)

    # Kreiranje direktorijuma za trenutnu klasu u train/test
    trening_klase_dir = os.path.join(trening_dir, ime_klase)
    test_klase_dir = os.path.join(test_dir, ime_klase)
    os.makedirs(trening_klase_dir, exist_ok=True)
    os.makedirs(test_klase_dir, exist_ok=True)

    # Premestanje slika u odgovarajuće direktorijume
    for putanja_slike in trening_slike:
        shutil.copy(putanja_slike, os.path.join(trening_klase_dir, os.path.basename(putanja_slike)))
    for putanja_slike in test_slike:
        shutil.copy(putanja_slike, os.path.join(test_klase_dir, os.path.basename(putanja_slike)))

    print(f"Klasa '{ime_klase}' podeljena: {len(trening_slike)} za trening, {len(test_slike)} za test.")

print("Podela završena! Trening i test podaci su organizovani u direktorijumu:", ciljni_dir)
```

Ovaj deo koda se koristi za pripremu podataka i postavke za obuku modela dubokog učenja. Evo šta se dešava u ovoj celiji:

Postavke: Definišu se osnovni parametri, uključujući veličinu ulaznih slika (input_shape) i broj klasa (num_classes).



**Priprema podataka:**


Generator podataka: Koristi se *ImageDataGenerator* za normalizaciju slika (skaliranje piksela na opseg [0, 1]). Zatim se koristi *flow_from_directory* da bi se učitao trening skup sa slikama, koje se automatski prebacuju u odgovarajući format sa odgovarajućim veličinama (224x224 piksela) i batch_size od 32. class_mode='categorical' označava da su klase predstavljene u obliku one-hot enkodiranja, a shuffle=True omogućava da se slike mešaju pre nego što se obučavaju.
**Prebrojavanje uzoraka po klasama: **

Prebrojavaju se uzorci po klasama koristeći train_generator.classes, što daje niz brojeva koji označavaju kojoj klasi svaka slika pripada.

**Izračunavanje težina klasa:**

Koristi se funkcija *compute_class_weight* da bi se izračunale težine za svaku klasu, uzimajući u obzir da li su klase uravnotežene (balansirane). Ove težine pomažu da se model ne favorizuje klase koje imaju veći broj uzoraka. Težine se zatim pretvaraju u rečnik za lakšu primenu prilikom obuke.

Ovaj proces omogućava da se podaci pripreme za obuku modela, uzimajući u obzir neravnotežu klasa i osiguravajući da model ne favorizuje većinske klase.

Ovaj deo koda se koristi za kreiranje modela dubokog učenja koristeći MobileNetV2 i dodavanje nekoliko dodatnih slojeva specifičnih za zadatak klasifikacije.

MobileNetV2 je konvoluciona neuronska mreža koja je optimizovana za uređaje sa ograničenim resursima. Ovdje se koristi sa ImageNet težinama (weights='imagenet'), što znači da je model već obučen na velikom skupu slika.
Parametar include_top=False znači da se ne uključuju poslednji slojevi predtrenirane mreže koji služe za klasifikaciju (tj. slojevi koji su specifični za ImageNet klase). Umesto toga, koristićemo naše vlastite slojeve za klasifikaciju.
Parametar input_shape=(224, 224, 3) označava da se očekuju ulazne slike dimenzija 224x224 piksela sa 3 kanala (RGB).
Zamrzavanje predtrenirane mreže:

base_model.trainable = False zamrzava slojeve MobileNetV2, što znači da se njihova težina neće menjati tokom obuke. Ovaj korak je važan jer želimo da iskoristimo već naučene osobine mreže bez potrebe za dodatnim učenjem tih slojeva.


**Dodavanje vlastitih slojeva:**


*GlobalAveragePooling2D:* Ovaj sloj se koristi da bi se izvršilo globalno prosečno grupisanje iznad poslednjih konvolucionih slojeva. Ovo pomaže da se smanji broj parametara i spreči prekomerno učenje.

*Dense(256, activation='relu'):* Dodaje potpuno povezani sloj sa 256 neurona i ReLU aktivacijom, što omogućava modelu da uči složene reprezentacije.

*Dropout(0.5):* Dodaje sloj za regularizaciju (dropout) sa verovatnoćom 0.5, što znači da će pola neurona biti isključeno tokom obuke, čime se sprečava prekomerno učenje.

*Dense(len(imena_klasa), activation='softmax'):* Na kraju, koristi se potpuno povezani sloj sa brojem neurona jednakim broju klasa (len(imena_klasa)). Aktivacija softmax omogućava da se izlaz modela interpretira kao verovatnoće za svaku klasu.

Ovaj model koristi predtrenirane slojeve MobileNetV2 kao osnovu i dodaje slojeve specifične za naš zadatak klasifikacije, omogućavajući efikasno prepoznavanje objekata sa manje podataka za obuku.



```python
from tensorflow.keras.applications import MobileNetV2

base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Zamrzni slojeve

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(len(imena_klasa), activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',  # Možeš zameniti sa SGD ili RMSprop, zavisno od problema
    loss='categorical_crossentropy',  # Koristi ovu funkciju jer imaš više klasa
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_generator,
    epochs=10,  # Broj epoha
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=5,
            verbose=1
        ),
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True
        )
    ]
)
```

```python
test_dir = "./skup_podataka/test"  # Putanja do test skupa

# Priprema generatora za test
test_datagen = ImageDataGenerator(rescale=1.0 / 255.0)  # Samo normalizacija

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False  # Ne mešamo test podatke
)
```

```python
test_loss, test_accuracy = model.evaluate(test_generator)
print(f"Test gubitak: {test_loss:.4f}")
print(f"Test tačnost: {test_accuracy:.4f}")
```

```python
sortirani_nazivi_klasa = sorted(os.listdir(test_dir))

# Učitavanje test slika
test_slike = []
test_oznake = []

# Prolazak kroz test direktorijum i učitavanje slika
for idx, ime_klase in enumerate(sortirani_nazivi_klasa):
    putanja_klase = os.path.join(test_dir, ime_klase)
    for fajl in os.listdir(putanja_klase):
        putanja_slike = os.path.join(putanja_klase, fajl)
        slika = load_img(putanja_slike, target_size=(224, 224))  # Prilagođavanje veličine slike
        niz_slika = img_to_array(slika) / 255.0          # Normalizacija [0, 1]
        test_slike.append(niz_slika)
        test_oznake.append(idx)

# Konvertovanje u numpy nizove
test_slike = np.array(test_slike)
test_oznake = np.array(test_oznake)

# Nasumični izbor 10 slika za testiranje
indeksi = np.random.choice(range(len(test_slike)), 10, replace=False)

# Inicijalizacija evidencije za tačnost
tacne_predikcije = 0
ukupno_predikcija = 0

# Testiranje modela
for i in indeksi:
    img = test_slike[i]
    prava_oznaka = imena_klasa[test_oznake[i]]

    # Proširivanje dimenzije slike za predikciju
    img_niz = np.expand_dims(img, axis=0)
    predikcija = model.predict(img_niz)
    indeks_predvidjene_klase = np.argmax(predikcija)
    predvidjena_klasa = imena_klasa[indeks_predvidjene_klase]

    # Provera da li je predviđena klasa tačna
    tacno = prava_oznaka == predvidjena_klasa
    if tacno:
        tacne_predikcije += 1
    ukupno_predikcija += 1

    # Prikaz slike sa stvarnom i predviđenom klasom
    plt.figure(figsize=(5, 6))
    plt.imshow(img)
    plt.title(f"Stvarna klasa: {prava_oznaka}\nPredviđena klasa: {predvidjena_klasa}\n{'Tačno' if tacno else 'Pogrešno'}")
    plt.axis('off')
    plt.show()

# Izračunavanje tačnosti modela
preciznost_predikcije = (tacne_predikcije / ukupno_predikcija) * 100
print(f"Procenat tačnosti: {preciznost_predikcije:.2f}%")
```
